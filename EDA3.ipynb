```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Load the dataset
data = pd.read_csv('/content/amazon (1).csv')
```

```python
data
```

```python

# 2. Top rated products by category
top_rated_products = data.loc[data.groupby('category')['rating_count'].idxmax()]
print(top_rated_products[['category', 'product_name', 'rating_count']])

# 3. Distribution of discounted prices vs. actual prices
plt.figure(figsize=(10, 6))
sns.histplot(data['discounted_price'], color='blue', label='Discounted Price', kde=True)
sns.histplot(data['actual_price'], color='red', label='Actual Price', kde=True)
plt.legend()
plt.title('Distribution of Discounted vs Actual Prices')
plt.show()

#4. Average discount percentage across categories
#avg_discount_per_category = data.groupby('category')['discount_percentage'].mean().reset_index()
#print(avg_discount_per_category)

# 5. Most popular product names
popular_products = data['product_name'].value_counts().head(10)
print(popular_products)

# 6. Most popular product keywords
from collections import Counter
keywords = ' '.join(data['about_product'].dropna()).split()
popular_keywords = Counter(keywords).most_common(10)
print(popular_keywords)

# 7. Most popular product reviews
popular_reviews = data['review_content'].value_counts().head(10)
print(popular_reviews)

# 8.Correlation between discounted price and rating
#correlation = data[['discounted_price', 'rating']].corr()
#print(correlation)
```

```python
# 1. Average rating for each product category

avg_rating_per_category = data.groupby('category')['rating'].mean().reset_index()
print(avg_rating_per_category)


#4. Average discount percentage across categories
avg_discount_per_category = data.groupby('category')['discount_percentage'].mean().reset_index()
print(avg_discount_per_category)



# 8.Correlation between discounted price and rating
correlation = data[['discounted_price', 'rating']].corr()
print(correlation)
```
